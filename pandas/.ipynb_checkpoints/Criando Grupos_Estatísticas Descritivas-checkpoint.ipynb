## Criando Grupos

```python
import pandas as pd
dados = pd.read_csv('aluguel_residencial.csv', sep=';')
dados.head()
```

```python
dados['Valor'].mean()
```

```python
dados['Bairro'].drop_duplicates()
```

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
for bairro, data in grupo_bairro: 
    print(bairro)
```

```python
for bairro, data in grupo_bairro: 
    print('Bairro: {} -> Média do valor: {}'.format(bairro, data.Valor.mean()))
```

```python
grupo_bairro['Valor'].mean().round(2)
```

```python
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

## Estatísticas Descritivas

```python
grupo_bairro['Valor'].describe().round(2)
```

```python
grupo_bairro['Valor'].aggregate(['mean','min','max']).rename(columns = {'mean': 'Média','min':'Mínimo','max':'Máximo'})
```

## Gráficos

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

### Valor

```python
fig_std = grupo_bairro['Valor'].std().plot.bar()
fig_std.set_title('Desvio Padrão',{'fontsize':22})
```

```python
fig_mean = grupo_bairro['Valor'].mean().plot.bar()
fig_mean.set_title('Valor médio do aluguel por Bairro',{'fontsize':22})
```

```python
fig_max = grupo_bairro['Valor'].max().plot.bar()
fig_max.set_title('Valor Máximo',{'fontsize':22})
```

Analisando esses gráficos podemos perceber que tem algo de errado em nossa base de dados, provavelmente temos valores de venda de imóveis misturados com valores de aluguel, fazendo com que exista uma diferença gritante entre valores do alugel.

### IPTU

```python
fig_std = grupo_bairro['IPTU'].std().plot.bar()
fig_std.set_title('Desvio Padrão',{'fontsize':22})
```

```python
fig_mean = grupo_bairro['IPTU'].mean().plot.bar()
fig_mean.set_title('Valor médio do IPTU por Bairro',{'fontsize':22})
```

```python
fig_max = grupo_bairro['IPTU'].max().plot.bar()
fig_max.set_title('IPTU Máximo',{'fontsize':22})
```

Em nossa base de dados, temos muitas entradas com o valor do IPTU zerado, isso pode reprensentar um problema, os dados referentes ao IPTU não são confiáveis

## Análise por quantidade de quartos 

```python
grupo_quartos = dados.groupby('Quartos')
```

```python
dados['Quartos'].unique()
```

```python
fig_mean = grupo_quartos['Valor'].mean().plot.bar()
fig_mean.set_title('Valor médio do aluguel por quantidade de quartos',{'fontsize':22})
```
