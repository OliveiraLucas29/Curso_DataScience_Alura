```python
%matplotlib inline
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
dados = pd.read_csv('aluguel_amostra.csv', sep=';')
dados
```

```python
dados.boxplot('Valor m2')
```

```python
metro = dados['Valor m2']
```

```python
q1 = metro.quantile(.25)
q3 = metro.quantile(.75)
iiq = q3-q1
limite_inferior = q1-1.5*iqq
limite_superior = q3+1.5*iqq
```

```python
print("Q1:",q1)
print("Q3:",q3)
print("IIQ:",iiq.round(2))
print("Limite Inferior:",limite_inferior.round(2))
print("Limite Superior:",limite_superior.round(2))
```

```python
sel = (metro >= limite_inferior) & (metro <= limite_superior)
new_data = dados[sel]
new_data
```

```python
new_data.boxplot('Valor m2')
```
