a) Criar nova coluna com o valor bruto <br>
b) Criar nova coluna com o valor por metro quadrado <br>
c) Criar nova coluna que define o tipo como casa ou apartamento <br>

## Importação de bibliotecas e da base de dados

```python
import pandas as pd
dados = pd.read_csv("aluguel_residencial.csv", sep=';')
dados.head()
```

## Valor Bruto 

```python
dados['Valor_bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head()
```

## Valor por metro quadrado

```python
dados['Valor_por_metro'] = (dados['Valor']/dados['Area']).round(2)
dados.head()
```

### Valor Bruto por metro quadrado

```python
dados['Valor_bruto_por_metro'] = (dados['Valor_bruto']/dados['Area']).round(2)
dados.head()
```

## Casas e Apartamentos

```python
dados.Tipo.unique()
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo_agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados
```

# Excluindo Colunas

As colunas de valor bruto e valor bruto por metro quadrado não são muito confiáveis, pois muitos valores na coluna IPTU são iguais a zero, portando vamos excluir essas colunas

```python
dados.drop(['Valor_bruto_por_metro', 'Valor_bruto'], axis=1, inplace=True )
dados.head()
```

```python
dados.to_csv('aluguel_residencial.csv', sep=';',index=False)
```
