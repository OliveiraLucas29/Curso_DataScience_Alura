# Aula 1 - Primeiras Visualizações

## Importando bibliotecas e base de dados iniciais 

```python
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados = pd.read_csv('monitoramento_tempo.csv')
```

## Conhecendo nossa Base de Dados

```python
dados.head()
```

```python
dados.shape[0]
```

```python
dados.info()
```

A nossa variável data não é realmente do tipo data, isso pode nos atrapalhar futuramente, por isso vamos modificá-la.

```python
dados['data'] = pd.to_datetime(dados['data'])
```

```python
dados.info()
```

Agora sim podemos prosseguir!

## Primeiras Visualizações

```python
plt.plot(dados['data'], dados['temperatura'])
```

Parece estranho a primeira vista esses valores elevados da variável temperatura, isso acontece porque ela não está em graus celsius e sim em Kelvin

### Aumentando o tamanho do gráfico

```python
plt.figure(figsize = (11,5))
plt.plot(dados['data'], dados['temperatura'])
```

### Adicionando Título

```python
plt.figure(figsize = (11,5))
plt.plot(dados['data'], dados['temperatura'])
plt.title('Temperatura no decorrer do tempo')
```

## Funcionalidades básicas do Matplotlib

### Criando o gráfico

```python
fig = plt.figure(figsize=(11,5))
eixo = fig.add_axes([0,0,1,1])
```

```python
fig = plt.figure(figsize=(11,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'])
eixo.set_title('Temperatura no tempo')
```

O resultado final foi o mesmo, mas a criação das variáveis fig e eixo nos permitem fazer uma customização maior nos gráficos

### Adicionando labels

```python
fig = plt.figure(figsize=(11,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'])
eixo.set_title('Temperatura no tempo')
eixo.set_xlabel('Tempo (Anos)')
eixo.set_ylabel('Temperatura (K)')
```

### Ajustando tamanho da fonte

```python
fig = plt.figure(figsize=(11,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'])
eixo.set_title('Temperatura no tempo', fontsize = 17)
eixo.set_xlabel('Tempo (Anos)', fontsize = 13)
eixo.set_ylabel('Temperatura (K)', fontsize = 13)
```

### Adicionando legenda 

```python
fig = plt.figure(figsize=(11,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'])
eixo.set_title('Temperatura no tempo', fontsize = 17)
eixo.set_xlabel('Tempo (Anos)', fontsize = 13)
eixo.set_ylabel('Temperatura (K)', fontsize = 13)
eixo.legend(['Temperatura'], loc = 'upper left', fontsize = 12)
```

### Mudando a cor

```python
fig = plt.figure(figsize=(11,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'], color = 'lightgreen') # nome da cor em inglês
eixo.set_title('Temperatura no tempo', fontsize = 17)
eixo.set_xlabel('Tempo (Anos)', fontsize = 13)
eixo.set_ylabel('Temperatura (K)', fontsize = 13)
eixo.legend(['Temperatura'], loc = 'upper left', fontsize = 12)
```

```python
fig = plt.figure(figsize=(11,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'], color = '#088F8F') # código hex
eixo.set_title('Temperatura no tempo', fontsize = 17)
eixo.set_xlabel('Tempo (Anos)', fontsize = 13)
eixo.set_ylabel('Temperatura (K)', fontsize = 13)
eixo.legend(['Temperatura'], loc = 'upper left', fontsize = 12)
```

# Aula 2 - Customizando nossas visualizações

## Espessura, estilo, foco, marcadores e grades

### Mudando a espessura da linha do gráfico

```python
fig = plt.figure(figsize=(11,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'], color = '#088F8F', lw = 4) # lw = line weight
eixo.set_title('Temperatura no tempo', fontsize = 17)
eixo.set_xlabel('Tempo (Anos)', fontsize = 13)
eixo.set_ylabel('Temperatura (K)', fontsize = 13)
eixo.legend(['Temperatura'], loc = 'upper left', fontsize = 12)
```

```python
fig = plt.figure(figsize=(11,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'], color = '#088F8F', lw = 0.4) # lw = line weight
eixo.set_title('Temperatura no tempo', fontsize = 17)
eixo.set_xlabel('Tempo (Anos)', fontsize = 13)
eixo.set_ylabel('Temperatura (K)', fontsize = 13)
eixo.legend(['Temperatura'], loc = 'upper left', fontsize = 12)
```

### Mudando estilo da linha

```python
fig = plt.figure(figsize=(11,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'], color = '#088F8F', ls = 'dotted') # ls = line style
eixo.set_title('Temperatura no tempo', fontsize = 17)
eixo.set_xlabel('Tempo (Anos)', fontsize = 13)
eixo.set_ylabel('Temperatura (K)', fontsize = 13)
eixo.legend(['Temperatura'], loc = 'upper left', fontsize = 12)
```

### Visualizando período específico de tempo

```python
import datetime
```

```python
fig = plt.figure(figsize=(11,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'], color = '#088F8F') 
eixo.set_title('Temperatura no tempo', fontsize = 17)
eixo.set_xlabel('Tempo (Janeiro 2014 a Janeiro 2015)', fontsize = 13)
eixo.set_ylabel('Temperatura (K)', fontsize = 13)
eixo.legend(['Temperatura'], loc = 'upper left', fontsize = 12)
eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1)) # 01/01/2014 até 01/01/2015
```

### Adicionando Marcadores

```python
fig = plt.figure(figsize=(11,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'], color = '#088F8F', marker = 'o') 
eixo.set_title('Temperatura no tempo', fontsize = 17)
eixo.set_xlabel('Tempo (Janeiro 2014 a Janeiro 2015)', fontsize = 13)
eixo.set_ylabel('Temperatura (K)', fontsize = 13)
eixo.legend(['Temperatura'], loc = 'upper left', fontsize = 12)
eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
```

A utilização de marcadores pode ser muito útil dependendo do contexto. Em nosso caso mais atrapalha do que ajuda, por isso o ideal é não utilizar

### Criando grades

```python
fig = plt.figure(figsize=(11,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'], color = '#088F8F') 
eixo.set_title('Temperatura no tempo', fontsize = 17)
eixo.set_xlabel('Tempo (Janeiro 2014 a Janeiro 2015)', fontsize = 13)
eixo.set_ylabel('Temperatura (K)', fontsize = 13)
eixo.legend(['Temperatura'], loc = 'upper left', fontsize = 12)
eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo.grid(True)
```

## Mais de um eixo na mesma visualização

### colocando dados no segundo gráfico

```python
fig = plt.figure(figsize=(11,5))
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.5,0.4,0.4])

ax.grid(True)
ax.plot(dados['data'], dados['temperatura'], color = '#92A8D1')
ax.set_title('Temperatura em 2014', fontsize=17, pad = 12) 
ax.set_xlabel('Tempo (Janeiro 2014 até Janeiro 2015)', fontsize=13)
ax.set_ylabel('Temperatura (K)', fontsize=13)
ax.legend(['Temperatura'], loc = 'lower left', fontsize = 12)
ax.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))

ax2.grid(True)
ax2.plot(dados['data'], dados['temperatura'], color = '#009B77')
ax2.set_title('Temperatura no decorrer do tempo', fontsize=14, pad = 12) 
ax2.set_xlabel('Tempo (Anos)', fontsize=10)
ax2.set_ylabel('Temperatura (K)', fontsize=10)
ax2.legend(['Temperatura'], loc = 'best', fontsize = 8)
```

## Uma visualização mais complexa

Nosso objetivo agora é visualizar de maneira específica o mês de maio de 2014

```python
fig = plt.figure(figsize=(11,5))
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.7,0.66,0.29,0.29])

ax.grid(True)
ax.set_xlim(datetime.datetime(2014,5,1),datetime.datetime(2014,6,1))
ax.set_ylim(275,320)
ax.plot(dados['data'], dados['temperatura'], color = '#009B77')
ax.set_title('Temperatura em  Maio de 2014', fontsize=17, pad = 12) 
ax.set_xlabel('Tempo', fontsize=13)
ax.set_ylabel('Temperatura (K)', fontsize=13)
ax.legend(['Temperatura'], loc = 'lower left', fontsize = 12)

ax2.grid(True)
ax2.plot(dados['data'], dados['temperatura'], color = '#98B4D4')
ax2.set_title('Temperatura em 2014', fontsize=10) 
ax2.set_xlabel('Janeiro de 2014 até Dezembro 2014', fontsize=8)
ax2.set_ylabel('Temperatura (K)', fontsize=8)
ax2.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
```

### Colorindo o mês de maio em nosso gráfico do ano de 2014

```python
fig = plt.figure(figsize=(11,5))
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.68,0.65,0.29,0.29])

ax.grid(True)
ax.set_xlim(datetime.datetime(2014,5,1),datetime.datetime(2014,6,1))
ax.set_ylim(275,320)
ax.plot(dados['data'], dados['temperatura'], color = '#009B77')
ax.set_title('Temperatura em Maio de 2014', fontsize=17, pad = 12) 
ax.set_xlabel('Data', fontsize=13)
ax.set_ylabel('Temperatura (K)', fontsize=13)
ax.legend(['Temperatura'], loc = 'lower left', fontsize = 12)

azul_esquerda = dados['data'] < datetime.datetime(2014,5,1)
azul_direita = dados['data'] > datetime.datetime(2014,6,1)

ax2.grid(True)
ax2.plot(dados['data'], dados['temperatura'], color = '#009B77')
ax2.plot(dados[azul_esquerda]['data'],dados[azul_esquerda]['temperatura'], color = '#98B4D4')
ax2.plot(dados[azul_direita]['data'],dados[azul_direita]['temperatura'], color = '#98B4D4')
ax2.set_title('Temperatura em 2014', fontsize=10) 
ax2.set_xlabel('Janeiro de 2014 até Dezembro 2014', fontsize=8)
ax2.set_ylabel('Temperatura (K)', fontsize=8)
ax2.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
```
