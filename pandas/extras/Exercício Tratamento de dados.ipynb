# Exercício de Tratamento de dados

## Importando base de dados e pandas

```python
import pandas as pd
dados = pd.DataFrame([['Apartamento', None, 970, 68], 
                        ['Apartamento', 2000, 878, 112], 
                        ['Casa', 5000, None, 500], 
                        ['Apartamento', None, 1010, 170], 
                        ['Apartamento', 1500, 850, None], 
                        ['Casa', None, None, None], 
                        ['Apartamento', 2000, 878, None], 
                        ['Apartamento', 1550, None, 228], 
                        ['Apartamento', 2500, 880, 195]], 
                        columns = ['Tipo', 'Valor', 'Condominio', 'IPTU'])
dados
```

## 1) Removendo linhas onde a coluna Valor possui células nulas

```python
dados.dropna(subset='Valor', inplace=True)
dados
```

## 2) Removendo imóveis do tipo Apartamento que não apresentam valor Condominio

```python
apnul = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
dados = dados[~apnul]
dados
```

## 3) Substituindo valores nulos restantes por 0

```python
dados.fillna({'Condomínio':0,'IPTU':0})
dados
```

## 4) Arrumando o index

```python
dados.index=range(dados.shape[0])
dados
```
